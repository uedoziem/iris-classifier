# Iris Classifier (Decision Tree)

This notebook trains a Decision Tree classifier on the Iris dataset using scikit-learn.

## Steps:
1. Load the Iris dataset  
2. Split into train/test sets  
3. Train a Decision Tree classifier  
4. Evaluate accuracy  
5. Plot a confusion matrix  


## Importing Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

## Loading the iris dataset

```python
iris = load_iris()
X = iris.data
y = iris.target

print("Features shape:", X.shape)
print("Target shape:", y.shape)
print("Classes:", iris.target_names)
```

## Train/Test Split

```python
TEST_SIZE = 0.2
RANDOM_STATE = 42

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Train size:", X_train.shape[0])
print("Test size:", X_test.shape[0])
```

## Training Decision Tree

```python
model = DecisionTreeClassifier(random_state=RANDOM_STATE)
model.fit(X_train, y_train)

print("Model trained successfully.")
```

## Prediction + Accuracy

```python
y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f"Accuracy: {acc:.4f}")
```

## Plotting the Confusion Matrix

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    xticklabels=iris.target_names,
    yticklabels=iris.target_names
)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Decision Tree)")
plt.tight_layout()
plt.show()
```

## Results Interpretation

- The accuracy score shows the proportion of correct predictions on the test set.  
- The confusion matrix shows how well the model separates the three Iris classes:
  - Setosa is usually classified very accurately.
  - Most errors often occur between Versicolor and Virginica, because their measurements can overlap.
- Overall, the model performs well on this dataset using a simple Decision Tree classifier.
